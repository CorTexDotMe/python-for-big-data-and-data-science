15. Порівняйте список та кортеж. Назвіть схожості та відмінності, випадки використання.

Схожості між списком і кортежем:

- Обидва це колекції даних, які можуть зберігати декілька елементів.
- Обидва можуть зберігати різні типи даних в своїх елементах(хоч ідейно краще не робити цього у кортежах).
- Обидва підтримують доступ до елементів за допомогою індексів.

Відмінності між списком і кортежем:

- Списки є мутабельними, тобто їх можна змінювати (додавати, видаляти елементи), тоді як кортежі є немутабельними, тобто їхні елементи не можна змінювати після створення.
- Списки створюються за допомогою квадратних дужок [], а кортежі - за допомогою круглих дужок ().
-  Кортежі зазвичай працюють швидше, оскільки вони є немутабельними, що дозволяє оптимізувати деякі операції.

Випадки використання:

- Списки використовуються там, де потрібно зберігати дані, які можуть змінюватися. Наприклад, список користувачів, список товарів у кошику тощо.
- Кортежі зазвичай використовуються там, де потрібно зберігати статичні дані, які не повинні змінюватися. Наприклад, координати точки на площині, розмірність фігури, відомості про дні тижня тощо.

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
set1 = {1, 2, 3, 4, 5}
set2 = {5, 6, 7, 8, 9}

print(set1.union(set2))
print(set1.difference(set2))
print(set1.intersection(set2))
print(set1.symmetric_difference(set2))
```

# VII. Словники (Dictionary).

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
my_dict = {}

my_dict[1] = 1.0
my_dict["2"] = 2.0
my_dict[(3,3)] = 3.0
my_dict[4.0] = 4.0
print(my_dict)
```

Список не може бути ключем, тому що ключ має бути хешованим. В основі словників лежить ідея хеш-таблиць, де для кожного ключа потрібно мати хеш, щоб точно розрізняти ключі між собою. Список не є хешованим, тому що список можна змінити, а від цього зміниться хеш.
